## Example 1: Binomial Distribution

### 1. calculate the success rate

```julia
coin(p) = rand() < p
coins(N, p) = [coin(p) for i in 1:N]
success_rate(N, p) = sum(coins(N, p)) / N;
```

```julia
N = 100
p = 0.25
num_times = 1000

rates = [success_rate(N, p) for i in 1:num_times];
```

```julia
@manipulate for n in 1:1000
    histogram(rates[1:n], bins=100, normed=false)
end
```

### 2. catculate the sd of binomial distribution

```julia
Binom_var(p, N) = N*p*(1-p);
```

(1) Heatmap plots

```julia
combns = collect(Iterators.product(0:0.1:1, 1:10:500));
values = [Binom_var(unit[1], unit[2]) for unit in combns]; 
heatmap(values, xlabel="total number trials", ylabel="probability of successes",
    title="Variance in binomial distribution")
```

(2) 3D plot

```julia
combinations = collect(Iterators.product(0:0.1:1, 1:10:100))[:];
results_all = [[unit[1], unit[2], Binom_var(unit[1], unit[2])] for unit in combinations];
# concatenates horizontally and transpose
result_concat = hcat(results_all...)' ;
```

```julia
# number of bins need to be change by the intervals when you produce the data
@manipulate for p in 0:0.1:1, n in 1:10:100
    subdata = result_concat[(result_concat[:,1] .<= p) .& (result_concat[:,2] .<= n), :]
    plot(subdata[:, 1], subdata[:, 2], subdata[:, 3],
        xlabel = "probability of success", ylabel = "total number of trials",
        zlabel = "Variance", xlims=(0, 1), ylims = (1, 100), zlims = (0, 25),
    st=:surface)
end
```
